```python
hastags=[]
mentions=[]
name=[]
screen_name = []
ID=[]
posts=[]
time=[]
favourite=[]
retweet=[]

for mmmm_name in musician_screen_name:
    results = twitter_api.statuses.user_timeline(screen_name = mmmm_name)
    
    for status in results:
        str5 = '#'
        str6 = '@'
    
    #print(status['entities']['hashtags'])
        for seq in (status['entities']['hashtags']):
            #print(seq)
            str5 += seq['text'] + '#'
        hastags.append(str5)
        name.append(mmmm_name)
        ID.append(status['id'])
        posts.append(status['text'])
        favourite.append(status['favorite_count'])
        retweet.append(status['retweet_count'])
        time.append(status['created_at'])
    
        for seq in (status['entities']['user_mentions']):
            #print(seq)
            str6 += seq['screen_name'] + '@'
        mentions.append(str6)


```
